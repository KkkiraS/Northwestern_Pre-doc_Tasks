```python
def read_pdf(file_path):
    table = camelot.read_pdf(file_path, pages='all')
    df1 = pd.DataFrame(columns = [i for i in range(16)])
    for i in range(len(table)):
        df1 = pd.concat([df1, table[i].df]).reset_index(drop=True)
    df1.columns = df1.iloc[3]
    df1['date'] = df1.iloc[2,0]
    df1['type'] = df1.iloc[1,0]#.split(' \\ ')[1]
    df1.drop([0,1,2,3], inplace = True)
    df1.reset_index(inplace= True, drop = True)
    return df1
#         df1.append(table[i].df)
#     return df1
```

```python
final_df = pd.DataFrame(columns = [i for i in range(18)])
for file in os.listdir():
    file_path = "/".join([path,file])
    add_df = read_pdf(file_path)
    final_df = pd.concat([final_df, add_df]).reset_index(drop=True)
```
